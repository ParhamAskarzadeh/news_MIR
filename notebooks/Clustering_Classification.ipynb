```python
import numpy as np
import pandas as pd
```

```python
predicted_label = np.load("./data/cluster_label.npy")
classification_df = pd.read_csv('./data/classification.csv')
news_details = np.load("./data/news_details.npy", allow_pickle=True)  # Duplicate
```

```python
def classification(id):
    selected_row = classification_df.loc[classification_df['ids'] == id]
    return {'title': selected_row.iloc[0]['titles'], 'url': selected_row.iloc[0]['urls'], 'class/cluster':str(selected_row.iloc[0]['pred_labels'])}
```

```python
news_details[1]
```

```python
def clustering(id):
    title = None
    url = None
    cluster = None
    for index, detail_dict in np.ndenumerate(news_details):
        if id == detail_dict['id']:
            title = detail_dict['title']
            url = detail_dict['url']
            cluster = str(predicted_label[index])

    return {'title': title, 'url': url, 'class/cluster': cluster}
```

```python
def clustering_classification(id, cluster=True):
    if cluster:
        return clustering(id)
    try:
        return classification(id)
    except:
        return {'title': 'این خبر در مجموعه تست نیست', 'url': 'https://www.google.com/', 'class/cluster': 'ناموجود'}
```
