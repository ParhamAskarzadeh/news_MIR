```python
import json
import pandas as pd
import numpy as np
```

```python
pagerank_file = open('./data/page_rank_result')
pagerank = json.load(pagerank_file)
news_details = np.load("./data/news_details.npy", allow_pickle=True)  # Duplicate
```

```python
def PageRank(k):
    pr_list = []
    largest_hubs_indices = sorted(pagerank, key=pagerank.get, reverse=True)[:k]
    for i in largest_hubs_indices:
        pr_list.append({'title': news_details[int(i)]['title'], 'url': news_details[int(i)]['url']})

    return pr_list
```

```python
PageRank(10)
```
