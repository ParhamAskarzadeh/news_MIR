```python
import json
import pandas as pd
import numpy as np
```

```python
hits_hubs_file = open('./data/hits_hubs_result')
hits_authorities_file = open('./data/hits_auth_result')
hubs, authorities = json.load(hits_hubs_file), json.load(hits_authorities_file)
news_details = np.load("./data/news_details.npy", allow_pickle=True)  # Duplicate
```

```python
def HITS(k):
    hubs_list = []
    auths_list = []
    largest_hubs_indices = sorted(hubs, key=hubs.get, reverse=True)[:k]
    for i in largest_hubs_indices:
        hubs_list.append({'title': news_details[int(i)]['title'], 'url': news_details[int(i)]['url']})
    largest_authorities_indices = sorted(authorities, key=authorities.get, reverse=True)[:k]
    for i in largest_authorities_indices:
        auths_list.append({'title': news_details[int(i)]['title'], 'url': news_details[int(i)]['url']})

    return hubs_list, auths_list
```

```python
HITS(10)
```
